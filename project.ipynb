### 3. Simple plots (powered by matplotlib)

```python
serious_crimes = [round(df['Serious'][i:i + 12].mean()) for i in range(0, len(df), 12)]
drugs_crimes = [round(df['Drugs'][i:i + 12].mean()) for i in range(0, len(df), 12)]
theft_crimes = [round(df['Theft'][i:i + 12].mean()) for i in range(0, len(df), 12)]

date = [df['month'][i][6:] for i in range(0, len(df), 12)]

# Create a figure with subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Line plot
ax1.plot(date, serious_crimes, label='Serious Crimes', linewidth=3, color='red')
ax1.plot(date, drugs_crimes, label='Drug Crimes', linewidth=3, color='blue')
ax1.plot(date, theft_crimes, label='Theft Crimes', linewidth=3, color='purple')
ax1.set_title('Serious, Drug, and Theft Crimes Over Time')
ax1.set_ylabel('Mean Crimes')
ax1.legend()
ax1.grid(True)

# Scatter plot
ax2.scatter(date, serious_crimes, label='Serious Crimes', color='red')
ax2.scatter(date, drugs_crimes, label='Drug Crimes', color='blue')
ax2.scatter(date, theft_crimes, label='Theft Crimes', color='purple')
ax2.set_xlabel('Date')
ax2.set_ylabel('Mean Crimes')
ax2.legend()
ax2.grid(True)

# Adjust layout to prevent clipping of labels
plt.tight_layout()


# Create histograms
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.hist(serious_crimes, bins=10, color='red', alpha=0.7)
plt.title('Serious Crimes Histogram')
plt.xlabel('Mean Crimes')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(drugs_crimes, bins=10, color='blue', alpha=0.7)
plt.title('Drug Crimes Histogram')
plt.xlabel('Mean Crimes')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(theft_crimes, bins=10, color='purple', alpha=0.7)
plt.title('Theft Crimes Histogram')
plt.xlabel('Mean Crimes')
plt.ylabel('Frequency')

plt.tight_layout()
# Show the plot
plt.show()
```

### Cool plots with detailed overview (powered by plotly.express)

```python
import plotly.express as px

serious_crimes = df[df.columns[2:]].sum()
column_names = list(df)[2:]

fig = px.pie(df,
             values=serious_crimes, names=column_names,
             title='Percentage of the total number of crimes by type for 2003-2020',
             height=800, hole=0.1
            ).update_traces(textposition="inside", textinfo="percent+label")
fig.show()
```

```python
px.line(df[df.columns[2:]])
```

```python
df_filtered = df.drop(columns=['Total_crimes'])

df_filtered['month'] = pd.to_datetime(df_filtered['month'], format='%d.%m.%Y')

df_long = df_filtered.melt(id_vars=['month'], var_name='Crime_Type', value_name='Crime_Value')

df_long['Year'] = df_long['month'].dt.year

fig = px.line(df_long.groupby(['Year', 'Crime_Type']).mean().reset_index(), x='Year', y='Crime_Value', color='Crime_Type',
              title='Average value of crimes by year',
              labels={'Crime_Value': 'Average value of crimes', 'Year': 'Year', 'Crime_Type': 'Crime types'},
              width=1100, height=900, 
             markers=True)

fig.show()
```

```python
df['month'] = pd.to_datetime(df['month'], format='%d.%m.%Y')

# Get the year from the 'month' column
df['Year'] = df['month'].dt.year

# Convert data to long format
df_long = df.melt(id_vars=['Year', 'month'], value_vars=['Serious', 'Huge_damage', 'Theft', 'Drugs'],
                  var_name='Crime_Type', value_name='Crime_Value')

fig_hist = px.histogram(df_long, x='Year', y='Crime_Value', color='Crime_Type',
                        title='Average value of crimes by year',
                        labels={'Crime_Value': 'Average value of crimes', 'Year': 'Year', 'Crime_Type': 'Crime types'},
                        histfunc='avg')

fig_scatter = px.scatter(df_long, x='Year', y='Crime_Value', color='Crime_Type',
                          title='Average value of crimes by year',
                          labels={'Crime_Value': 'Average value of crimes by year', 'Year': 'Year', 'Crime_Type': 'Crime types'})

fig_hist.show()
fig_scatter.show()
```

```python
most_type_crimes = df[df.columns[2:]].idxmax(axis=1, numeric_only=1)
df['most crimes'] = most_type_crimes

least_type_crimes = df[df.columns[2:]].idxmin(axis=1, numeric_only=1)
df['most crimes'] = most_type_crimes
```

```python
import plotly.express as px

fig = px.line(df, x='month', y='Total_crimes', title='Total Crimes Over Time')
fig.show()
```

```python
fig = px.bar(df, x='month', y=['Serious', 'Huge_damage', 'Ecological', 'Terrorism', 'Extremism', 'Murder', 'Harm_to_health', 'Rape', 'Theft', 'Vehicle_theft', 'Fraud_scam', 'Hooligan', 'Drugs', 'Weapons'], title='Distribution of Different Types of Crimes')
fig.show()
```

```python
fig = px.area(df, x='month', y=['Serious', 'Huge_damage', 'Terrorism', 'Murder', 'Theft'], title='Trend of Major Crime Categories Over Time')
fig.show()
```

```python
correlation_matrix = df[df.columns[2:]].corr(method="spearman").round(3)

fig = px.imshow(correlation_matrix, x=correlation_matrix.index, y=correlation_matrix.columns,
                labels=dict(color="Correlation"), color_continuous_scale='Purples', text_auto=True, height=1000)

fig.update_layout(title='Correlation Between Different Types of Crimes')
fig.show()
```

```python
df['month'] = pd.to_datetime(df['month'])

# Extract the year from the 'month' column
df['year'] = df['month'].dt.year

averages_by_year = df.groupby('year').mean().reset_index()

averages_by_year.drop(['month', 'Total_crimes'], axis=1, inplace=True)

averages_by_year.to_csv('crime_averages_by_year.csv', index=False)
```
