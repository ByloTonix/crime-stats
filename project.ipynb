## Crime in Russia.

This dataset "Crime in Russia (2003-2020)" is taken from [kaggle.com](https://www.kaggle.com/datasets/tsarkov90/crime-in-russia-20032020).

It contains information about the state of crime in Russia since 2003.
The table shows the number of crimes of various types for each month.


Here is a basic information about the contents of the dataset:
```bash
Column Name      |  Description
====================================================================
Month            |  date in the format DD.MM.YYYY
Total_crimes     |  total number of crimes
Serious          |  number of serious crimes
Huge_damage      |  number of crimes which caused significant damage
Ecological       |  number of crimes with ecological problems
Terrorism        |  number of terroristic crimes
Extremism        |  number of extremistic crimes
Murder           |  number of reported murders
Harm_to_health   |  number of crimes causing harm to health
Rape             |  number of reported rape cases
Theft            |  number of reported theft cases
Vehicle_theft    |  number of reported vehicle theft cases.
Fraud_scam       |  number of fraud and scam cases.
Hooligan         |  number of hooliganism-related crimes.
Drugs            |  number of crimes related to drugs.
Weapons          |  number of crimes involving weapons.

```

### Simple plots (powered by matplotlib)

````python
serious_crimes = [round(df['Serious'][i:i + 12].mean()) for i in range(0, len(df), 12)]
drugs_crimes = [round(df['Drugs'][i:i + 12].mean()) for i in range(0, len(df), 12)]
theft_crimes = [round(df['Theft'][i:i + 12].mean()) for i in range(0, len(df), 12)]

date = [df['month'][i][6:] for i in range(0, len(df), 12)]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 6), sharex=True)

# Line plot
ax1.plot(date, serious_crimes, label='Serious Crimes', linewidth=3, color='red')
ax1.plot(date, drugs_crimes, label='Drug Crimes', linewidth=3, color='blue')
ax1.plot(date, theft_crimes, label='Theft Crimes', linewidth=3, color='green')
ax1.set_title('Serious, Drug, and Theft Crimes Over Time (Mean Values)')
ax1.set_ylabel('Values')
ax1.legend()
ax1.grid(True)

# Scatter plot
ax2.scatter(date, serious_crimes, label='Serious Crimes', color='red')
ax2.scatter(date, drugs_crimes, label='Drug Crimes', color='blue')
ax2.scatter(date, theft_crimes, label='Theft Crimes', color='green')
ax2.set_xlabel('Date')
ax2.set_ylabel('Values')
ax2.legend()
ax2.grid(True)

plt.tight_layout()


# Create histograms
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(serious_crimes, bins=10, color='red', alpha=0.7)
plt.title('Serious Crimes Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(drugs_crimes, bins=10, color='blue', alpha=0.7)
plt.title('Drug Crimes Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(theft_crimes, bins=10, color='green', alpha=0.7)
plt.title('Theft Crimes Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.tight_layout()
# Show the plot

date = [df['month'][i][6:] for i in range(14)]

plt.figure(figsize=(18, 7))

index = range(len(date))

plt.bar([i - 0.25 for i in index], serious_crimes[:14], color='blue', alpha=0.7, width=0.25, label='Serious Crimes')
plt.bar(index, drugs_crimes[:14], color='green', alpha=0.7, width=0.25, label='Drugs Crimes')
plt.bar([i + 0.25 for i in index], theft_crimes[:14], color='red', alpha=0.7, width=0.25, label='Theft Crimes')

plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Serious, Drug, and Theft Crimes Over Time (Mean Values)')
plt.xticks(index, date)
plt.legend()



plt.show()
````

Line charts, scatter and bar plots of Serious, Drug and Theft Crimes. First and last plots shows the level of crime for the certain year.

Using the first graph, we can conveniently view crime statistics by year and month.
Using the second graph, we can see that the total number of crimes was the highest in the period from 2003 to 2007, then it steadily decreased.

````python
correlation_matrix = df[df.columns[2:-1]].corr(method="spearman").round(3)

fig = px.imshow(correlation_matrix, x=correlation_matrix.index, y=correlation_matrix.columns,
                labels=dict(color="Correlation"), color_continuous_scale='Purples', text_auto=True, height=1000)

fig.update_layout(title='Correlation Between Different Types of Crimes')
fig.show()
````

According to the correlation matrix, Murder, Harm_To_health, Rape, Theft and Vehicle_theft crimes have the highest correlation, equal to almost the absolute value of unity. 

### Data transformation

Let's add some new columns and modify something:

````python
# Before
df.head(len(df))
````

````python
# After
averages_by_year = df.groupby('year').mean().reset_index()
averages_by_year.drop(['month', 'Total_crimes'], axis=1, inplace=True)
averages_by_year.to_csv('crime_averages_by_year.csv', index=False)

df['most crimes'] = df[df.columns[2:]].idxmax(axis=1, numeric_only=1)

df['least crimes'] = df[df.columns[2:]].idxmin(axis=1, numeric_only=1)

df.head(len(df))
````

Well, you could see that 'month' column has changed its data format earlier. Also, I've added year, most and least crimes columns. Moreover, I've exported new csv with mean values for year of every type of crime (it can be used, for example, to make a line plot of crimes by mean values).

Some visualisation for new information:

````python
most_crimes_count = df['most crimes'].value_counts().reset_index()
least_crimes_count = df['least crimes'].value_counts().reset_index()

plt.figure(figsize=(15, 6))
sns.countplot(x='most crimes', data=df, color='skyblue', label='Most')
sns.countplot(x='least crimes', data=df, color='salmon', label='Least')

plt.xlabel('Crime Type')
plt.ylabel('Value')
plt.title('Most and Least Types of Crimes By Month')
plt.legend()

plt.xticks(rotation=0)
plt.show()
````

We can see, that the "most popular" least type of crime is extremism, and "most popular" most type of crime is theft. Different reasons why this is so, but anyway.

### Hypothesis check

I assume that when the number of drug-related crimes increases, the number of weapon crimes also increases.
To prove or disprove this, we can use the previous correlation plot



But I'll calculate it again...

````python
drug_crimes = df['Drugs']
weapon_crimes = df['Weapons']

correlation_coefficient = drug_crimes.corr(weapon_crimes, method="spearman").round(3)

print(correlation_coefficient)
````

Correlation between Drugs and Weapon is less than 1 (0.232)

````python
fig = px.scatter(df, x='Drugs', y='Weapons', title='Correlation between Drugs and Weapons Crimes',
                 labels={'Drugs': 'Number of Drug Crimes', 'Weapons': 'Number of Weapons Crimes'},
                 trendline='ols')

fig.update_traces(marker=dict(size=12, color='rgba(0, 0, 255, 0.7)', line=dict(color='black', width=2)))
fig.update_layout(title=dict(font=dict(size=20, color='black')))

fig.show()
````

The graph shows a positive, but a very-very small correlation between drug-related crimes and weapon-related crimes. The trend line is pointing upwards. There are some points that deviate greatly from the trend line.

All this leads to the conclusion that drug-related crimes and weapon-related crimes are independent of each other.
